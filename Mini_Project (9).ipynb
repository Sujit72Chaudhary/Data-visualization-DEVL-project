# Brief description of the dataset

#DisNo.: A unique identifier for each disaster event. This helps in tracking and referencing specific disasters within the dataset.

Historic: A flag indicating whether the disaster event is part of historical records or data. "Yes" or "No" values are typically used.

Classification Key: A code or classification system used to categorize the type of disaster event. It could represent various aspects, like the severity or specific nature of the disaster.

Disaster Group: Broad category or classification of the disaster, such as "Natural" (e.g., earthquakes, floods) or "Technological" (e.g., industrial accidents, nuclear incidents).

Disaster Subgroup: A more specific category within the disaster group. For example, under "Natural," there could be subgroups like "Meteorological," "Geophysical," etc.

Disaster Type: The specific type of disaster under the subgroup, such as "Storm," "Flood," "Earthquake," or "Extreme Temperature." It provides more granular classification.

Disaster Subtype: A further breakdown of the disaster type, such as "Cold wave" or "Drought" under the "Extreme Temperature" type, or "Flood" and "Tsunami" under the "Flood" type.

External IDs: External identifiers that may refer to disaster records in other databases or systems, helping to cross-reference data between platforms.

Event Name: The formal name given to the disaster event, which may include location and date (e.g., "Hurricane Katrina" or "Great East Japan Earthquake").

ISO: The ISO country code for the country where the disaster occurred (e.g., "US" for the United States, "IND" for India).

Country: The full name of the country where the disaster took place (e.g., "United States," "India").

Subregion: The subregion within the country where the disaster occurred. It could refer to a province, state, or region depending on the country (e.g., "California" in the United States).

Region: A broader geographical classification of the affected area, such as "Asia," "Africa," or "North America."

Location: The specific location of the disaster, which could be a city, province, or other area names. It may also include coordinates or more detailed descriptions.

Origin: The cause or origin of the disaster. This could describe whether it was natural, human-made, or technological (e.g., "earthquake," "industrial accident").

Associated Types: Other disaster types that may be associated with the event. For example, a flood might also be associated with landslides or storms.

OFDA/BHA Response: Indicates if the U.S. Office of Foreign Disaster Assistance (OFDA) or Bureau for Humanitarian Assistance (BHA) responded to the disaster, showing involvement in relief efforts.

Appeal: Whether an international appeal for aid was made after the disaster. This is important for tracking humanitarian efforts and responses.

Declaration: Whether an official disaster or emergency declaration was made, which often triggers relief efforts or international assistance.

AID Contribution ('000 US$): The amount of financial aid (in thousands of U.S. dollars) that was contributed toward disaster relief, including government and international donations.

Magnitude: The strength or intensity of the disaster. For natural events like earthquakes or storms, this is the measured magnitude (e.g., 7.0 on the Richter scale for an earthquake).

Magnitude Scale: The scale used to measure the magnitude of the disaster. For example, the Richter scale for earthquakes or the Saffir-Simpson scale for hurricanes.

Latitude: The latitude coordinate of the disaster location, which helps pinpoint the geographic area of the event.

Longitude: The longitude coordinate of the disaster location, used together with latitude to map the disaster’s location on a global scale.

River Basin: The river basin or watershed affected by the disaster, particularly relevant for floods or river-related events.

Start Year: The year the disaster started, marking the beginning of the event.

Start Month: The month the disaster started, indicating the time frame of the event.

Start Day: The exact day the disaster began, providing the precise start date.

End Year: The year the disaster ended, marking the end of the event.

End Month: The month the disaster ended, useful for understanding the duration of the event.

End Day: The specific day the disaster concluded, helping to determine the event’s duration.

Total Deaths: The total number of confirmed deaths as a result of the disaster.

No. Injured: The number of people injured or harmed by the disaster.

No. Affected: The total number of people affected by the disaster, including the injured, homeless, or displaced.

No. Homeless: The number of people who were rendered homeless due to the disaster, often relevant in cases of floods, storms, and earthquakes.

Total Affected: The total number of people who were impacted in some way by the disaster, encompassing all individuals who experienced displacement, injury, or loss.

Reconstruction Costs ('000 US$): The estimated cost, in thousands of US dollars, required to rebuild or repair infrastructure and communities affected by the disaster.

Reconstruction Costs, Adjusted ('000 US$): The adjusted reconstruction cost, factoring in inflation, to give a more accurate measure of financial requirements over time.

Insured Damage ('000 US$): The portion of the total damage that is covered by insurance, in thousands of U.S. dollars.

Insured Damage, Adjusted ('000 US$): The inflation-adjusted insured damage figure, representing the real value of insured losses.

Total Damage ('000 US$): The total monetary value of all the damage caused by the disaster, in thousands of U.S. dollars.

Total Damage, Adjusted ('000 US$): The inflation-adjusted total damage cost, offering a more realistic understanding of the disaster's economic impact over time.

CPI: The Consumer Price Index used to adjust monetary values for inflation, helping standardize financial figures in the dataset.

Admin Units: Administrative regions (e.g., states, provinces) affected by the disaster. It can include detailed hierarchical administrative divisions within the country.

Entry Date: The date when the disaster record was entered into the dataset, marking the beginning of the data collection.

Last Update: The date when the disaster record was last updated or revised in the dataset, helping track the timeliness of the data.

#Before going to the obseravtion part first of all let's know these things:
Strength of Correlation:

0.7 to 1 (or -0.7 to -1): Strong correlation.
0.5 to 0.7 (or -0.5 to -0.7): Moderate correlation.
0.3 to 0.5 (or -0.3 to -0.5): Weak correlation.
0 to 0.3 (or -0.3 to 0): Very weak or no correlation.

Example Interpretation:
Suppose you have Total Deaths and Total Damage ('000 US$) with a correlation of 0.8. This suggests a strong positive relationship, indicating that higher death tolls are associated with higher economic damage.
If Total Deaths and Magnitude show -0.4, there’s a weak negative correlation, meaning that as Magnitude increases, Total Deaths tend to slightly decrease.

From the table, we can see "Total Deaths" and "No. Homeless" has a moderate correlation of 0.58, that shows disasters that result in higher fatalities also tend to cause higher numbers of homeless people.
Similarly, "Total Deaths" and "No. Affected" has also a moderate correlation of 0.44, that shows disasters that result in higher fatalities means more
number of people affected.

```python
import matplotlib.pyplot as plt

# Get disaster counts and calculate percentages
disaster_counts = df['Disaster Type'].value_counts()
labels = disaster_counts.index
sizes = disaster_counts.values
total = sizes.sum()

# Create custom labels with counts and percentages
legend_labels = [f"{label}: {size} ({(size / total) * 100:.2f}%)" for label, size in zip(labels, sizes)]

# Create an explode list (optional)
explode = [0.1] * len(labels)  # Slightly explode all slices

# Plot the pie chart without percentages on slices
fig, ax = plt.subplots(figsize=(8, 8))
wedges, texts = ax.pie(
    disaster_counts, 
    startangle=90, 
    explode=explode, 
    wedgeprops={'linewidth': 1, 'edgecolor': 'black'}, 
    textprops={'fontsize': 10}  # Adjust text size
)

# Adjust layout to avoid clipping
plt.tight_layout()
plt.title('Distribution of Disaster Types')

# Add custom legend with counts and percentages
ax.legend(wedges, legend_labels, title="Disaster Types", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Filter the country
India_data = df[df['Country'] == 'India']

# Group the following attributes
disaster_counts = India_data.groupby(['Start Year', 'Disaster Type']).size().unstack(fill_value=0)

# Choose a type of disaster to be modelized
storm_counts = disaster_counts['Storm']

# Data preprocessing for LR 
X = storm_counts.index.values.reshape(-1, 1) 
y = storm_counts.values 

# Create a Linear Regression Model
model = LinearRegression()
model.fit(X, y)

# Set the years to be predicted
future_years = np.array([[2024], [2025], [2026], [2027], [2028]])  # Can be re-adjusted here
predictions = model.predict(future_years)

# Visualize Linear Regression Model
plt.figure(figsize=(8, 6))
plt.scatter(X, y, color='blue', label='Historical Storm Counts')
plt.plot(X, model.predict(X), color='green', label='Linear Regression Line')
plt.scatter(future_years, predictions, color='red', label='Predicted Storm Counts', marker='x')
plt.title('Storm Counts and Linear Regression Model')
plt.xlabel('Year')
plt.ylabel('Number of Storms')
plt.legend()
plt.grid()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Show prediction
for year, prediction in zip(future_years.flatten(), predictions):
    print(f'Predicted number of storms in {year}: {int(prediction)}')
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
import numpy as np

# Assuming df is your cleaned dataset
df_clean = df.dropna(subset=['Total Affected', "Total Damage, Adjusted ('000 US$)", 'No. Injured', 'No. Homeless', 'Total Deaths'])

# Select features and target variable
X_clean = df_clean[['Total Affected', "Total Damage, Adjusted ('000 US$)", 'No. Injured', 'No. Homeless']]
y_clean = df_clean['Total Deaths']

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_clean, y_clean, test_size=0.2, random_state=42)

# Create and train the linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Predictions on test set
y_pred = model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error: {mse}')
print(f'R^2 Score: {r2}')

# Visualize the predicted vs actual values
plt.scatter(y_test, y_pred)
plt.plot([y_clean.min(), y_clean.max()], [y_clean.min(), y_clean.max()], 'k--', lw=2)
plt.xlabel('Actual Total Deaths')
plt.ylabel('Predicted Total Deaths')
plt.title('Linear Regression: Actual vs. Predicted Total Deaths')
plt.show()

# Predicting for the next 5 years
# You need to assume or estimate values for features for the next 5 years
future_years = [2024, 2025, 2026, 2027, 2028]
future_data = pd.DataFrame({
    'Total Affected': [50000, 55000, 60000, 65000, 70000],  # Example estimated values for 'Total Affected'
    "Total Damage, Adjusted ('000 US$)": [2000000, 2200000, 2400000, 2600000, 2800000],  # Example estimated values for 'Total Damage'
    'No. Injured': [1000, 1100, 1200, 1300, 1400],  # Example estimated values for 'No. Injured'
    'No. Homeless': [3000, 3200, 3400, 3600, 3800]  # Example estimated values for 'No. Homeless'
})

# Predict the total deaths for the next 5 years
future_predictions = model.predict(future_data)

# Show predicted deaths for the next 5 years
for year, pred in zip(future_years, future_predictions):
    print(f"Predicted total deaths in {year}: {pred:.0f}")

# Visualize the predictions over the years
plt.plot(future_years, future_predictions, marker='o', linestyle='-', color='b', label='Predicted Total Deaths')
plt.xlabel('Year')
plt.ylabel('Predicted Total Deaths')
plt.title('Predicted Total Deaths for the Next 5 Years')
plt.xticks(future_years)
plt.grid(True)
plt.legend()
plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report
import matplotlib.pyplot as plt

# Assuming df is your cleaned dataset
df_clean = df.dropna(subset=['Total Affected', "Total Damage, Adjusted ('000 US$)", 'No. Injured', 'No. Homeless', 'Total Deaths'])

# Select features
X_clean = df_clean[['Total Affected', "Total Damage, Adjusted ('000 US$)", 'No. Injured', 'No. Homeless']]

# Create a binary target variable for "High" vs "Low" total deaths
threshold = df_clean['Total Deaths'].median()  # Use median as threshold
y_clean = df_clean['Total Deaths'] > threshold  # True if 'Total Deaths' is greater than the threshold

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_clean, y_clean, test_size=0.2, random_state=42)

# Create and train the logistic regression model
model = LogisticRegression()
model.fit(X_train, y_train)

# Predictions on test set
y_pred = model.predict(X_test)

# Evaluate the model
print(f"Confusion Matrix:\n{confusion_matrix(y_test, y_pred)}\n")
print(f"Classification Report:\n{classification_report(y_test, y_pred)}\n")

# Visualizing the ROC curve
from sklearn.metrics import roc_curve, auc

fpr, tpr, thresholds = roc_curve(y_test, model.predict_proba(X_test)[:, 1])
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic Curve')
plt.legend(loc='lower right')
plt.show()

# Displaying the predicted probabilities (optional)
pred_prob = model.predict_proba(X_test)[:, 1]
plt.figure(figsize=(8, 6))
plt.hist(pred_prob, bins=10, color='lightblue', edgecolor='black')
plt.xlabel('Predicted Probability')
plt.ylabel('Frequency')
plt.title('Predicted Probability Distribution of "High" Deaths')
plt.show()
```
